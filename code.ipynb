```python
import random
import time
import matplotlib.pyplot as plt
import gradio as gr
import os

# Create a directory for images if it doesn't exist
if not os.path.exists("images"):
    os.makedirs("images")

# Simulated "vote ledger" to store valid votes
vote_ledger = {
    1: 0,  # Candidate 1 votes
    2: 0,  # Candidate 2 votes
    3: 0,  # Candidate 3 votes
    4: 0,  # Candidate 4 votes
    5: 0   # Candidate 5 votes
}

# Step 1: Setup Phase (Key Generation)
def zk_snark_setup():
    start_time = time.time()

    # Simulate random setup time between 0.2 and 1.5 seconds
    simulated_time = random.uniform(0.2, 1.5)
    time.sleep(simulated_time)  # Simulate setup process

    setup_time = time.time() - start_time
    return setup_time


# Step 2: Proving Phase (Generate Proof)
def zk_snark_prove(vote_value):
    start_time = time.time()

    # Simulate random proof generation time between 0.1 and 0.8 seconds
    simulated_time = random.uniform(0.1, 0.8)
    time.sleep(simulated_time)  # Simulate proving process

    proof_time = time.time() - start_time

    # Simulate proof size between 100 and 500 KB
    proof_size = random.uniform(100, 500)
    return proof_time, proof_size


# Step 3: Verification Phase (Verify Proof)
def zk_snark_verify(vote_value):
    start_time = time.time()

    # Simulate random verification time between 0.05 and 0.5 seconds
    simulated_time = random.uniform(0.05, 0.5)
    time.sleep(simulated_time)  # Simulate verification process

    # Simulate verification result based on whether the vote is within a valid range
    verification_result = (1 <= vote_value <= 5)  # Example: valid vote if between 1 and 5
    verification_time = time.time() - start_time
    return verification_result, verification_time


# Function to record the vote in the vote ledger
def record_vote(vote_value):
    if 1 <= vote_value <= 5:
        vote_ledger[vote_value] += 1  # Increment the vote count for the chosen candidate
        return f"Vote recorded for Candidate {vote_value}."
    else:
        return "Invalid vote."


# Combine zk-SNARK simulation steps
def run_zk_snark(vote_value):
    setup_time = zk_snark_setup()    # Generate the public/private keys
    proof_time, proof_size = zk_snark_prove(vote_value)  # Prove possession of vote
    verification_result, verification_time = zk_snark_verify(vote_value)   # Verify the proof

    return setup_time, proof_time, proof_size, verification_time, verification_result


# Function to create and save visualizations
def visualize_results(setup_time, proof_time, proof_size, verification_time):
    fig, axs = plt.subplots(2, 2, figsize=(10, 10))

    # Setup Time Visualization
    axs[0, 0].bar(['Setup Time'], [setup_time], color='blue')
    axs[0, 0].set_title('Setup Time (seconds)')

    # Proof Size Visualization
    axs[0, 1].bar(['Proof Size'], [proof_size], color='orange')
    axs[0, 1].set_title('Proof Size (KB)')

    # Proof Generation Time Visualization
    axs[1, 0].bar(['Proof Time'], [proof_time], color='green')
    axs[1, 0].set_title('Proof Generation Time (seconds)')

    # Verification Time Visualization
    axs[1, 1].bar(['Verification Time'], [verification_time], color='red')
    axs[1, 1].set_title('Verification Time (seconds)')

    plt.tight_layout()

    # Save the plot as an image
    image_path = "images/zk_snark_voting_visualization.png"
    plt.savefig(image_path)
    plt.close()

    return image_path


# Gradio Interface
def zk_snark_voting_interface(vote_value):
    # Run the zk-SNARK simulation
    setup_time, proof_time, proof_size, verification_time, verification_result = run_zk_snark(vote_value)

    # Generate the visualizations and get the image path
    image_path = visualize_results(setup_time, proof_time, proof_size, verification_time)

    # Output verification result
    if verification_result:
        # If the vote is valid, record the vote
        vote_message = record_vote(vote_value)
        result = f"Vote successfully verified! {vote_message}"
    else:
        result = f"Vote verification failed. Invalid vote choice: {vote_value}. Please vote between 1 and 5."

    return result, image_path


# Function to display the current vote ledger (vote counts for each candidate)
def display_vote_ledger():
    ledger_message = "\n".join([f"Candidate {candidate}: {votes} votes" for candidate, votes in vote_ledger.items()])
    return ledger_message


# Real-world use case description
description = """
### Blockchain Voting Simulation Using zk-SNARKs

In this simulation, we demonstrate how zk-SNARKs can be used to ensure voter privacy in a blockchain voting system. Voters can cast their vote by entering a number representing their candidate choice.

#### How it works:
1. You submit your vote by choosing a number between 1 and 5, representing your candidate.
2. The zk-SNARK system generates a cryptographic proof that you cast your vote without revealing your actual vote.
3. The system verifies the proof to ensure the vote is valid, without knowing which candidate you voted for.
4. The vote is saved in the vote ledger, simulating a blockchain recording valid votes while preserving privacy.

#### Instructions:
- Input a number between 1 and 5 to cast your vote.
- The system will simulate the setup, proof generation, and verification phases of a zk-SNARK-based voting system.
- If your vote is valid, it will be successfully verified, and your vote will be saved in the ledger.
- You can also check the current vote ledger to see the vote count for each candidate.
"""

# Gradio Input and Output UI
gr_interface = gr.Interface(
    fn=zk_snark_voting_interface,
    inputs="number",  # Input: secret vote (number representing candidate)
    outputs=["text", "image"],  # Output: verification result and the image visualization
    title="Blockchain Voting Simulation Using zk-SNARKs",
    description=description,
    live=True
)

# Add a button to display the current vote ledger
vote_ledger_interface = gr.Interface(
    fn=display_vote_ledger,
    inputs=[],
    outputs="text",
    title="Current Vote Ledger",
    description=description
)

# Launch the Gradio UI with both interfaces
gr.Interface(
    fn=zk_snark_voting_interface,
    inputs="number",
    outputs=["text", "image"],
    live=True,
    description=description
).launch(share=True)

vote_ledger_interface.launch(share=True)
```
